## Missing value Treatement

***The data in df(penguins_lter) shows the missing values in Culmen Length (mm),Culmen Depth (mm),Flipper Length (mm),Body Mass (g),Sex,Delta 15 N (o/oo),Delta 13 C (o/oo) & Comments.***

*we can replace with previous values for Culmen Length (mm),Culmen Depth (mm),Flipper Length (mm),Body Mass (g) as they have only 2 missing values*


*For Sex column we can replace it with 0 as we dont know the sex as well as cant delete the rows.*


*We will replace the Delta 15 N (o/oo),Delta 13 C (o/oo) with mean and as comments column had major missing values and not usable feature so we will drop the column*

```python
df["Culmen Length (mm)"].fillna( method ='pad', inplace = True)
```

```python
df["Culmen Depth (mm)"].fillna( method ='pad', inplace = True)
```

```python
df["Flipper Length (mm)"].fillna( method ='pad', inplace = True)
```

```python
df["Body Mass (g)"].fillna( method ='pad', inplace = True)
```

```python
df.head()
```

```python
df["Sex"].fillna( value ='No Sex', inplace = True)
```

```python
mean_value = df["Delta 15 N (o/oo)"].mean() # Replacing with mean values
mean_value1 = df["Delta 13 C (o/oo)"].mean()
df["Delta 15 N (o/oo)"].fillna( value =mean_value, inplace = True)
df["Delta 13 C (o/oo)"].fillna( value =mean_value1, inplace = True)
df.head()
```

```python
df1 = df.drop(['Comments'], axis=1) #Dropping comments column
```
